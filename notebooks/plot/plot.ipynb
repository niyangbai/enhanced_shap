```python
import pickle
from shap_enhanced.tools.visulization import plot_mse_pearson, plot_3d_bars, plot_feature_comparison, plot_3d_surface
```

```python
with open("seq.pkl", "rb") as f:
    cache = pickle.load(f)

plot_mse_pearson(cache["results"], cache["pearson_results"], save='seq_mse_pearson.pdf')
# plot_3d_bars(cache["shap_gt"], cache["shap_models"], cache["seq_len"], cache["n_features"])
# plot_3d_surface(cache["shap_gt"], cache["shap_models"], cache["seq_len"], cache["n_features"])

for name, arr in cache["shap_models"].items():
    plot_3d_surface(
        cache["shap_gt"],
        {name: arr},  # single model as dict
        cache["seq_len"],
        cache["n_features"],
        save=f"seq_{name}.pdf",
        show=False
    )
    print(f"Saved: seq_{name}.pdf")
```

```python
from itertools import islice

items = list(cache["shap_models"].items())
batch_size = 4

for i in range(0, len(items), batch_size):
    batch = items[i:i+batch_size]
    batch_dict = {k: v for k, v in batch}
    plot_3d_surface(
        cache["shap_gt"],
        batch_dict,
        cache["seq_len"],
        cache["n_features"],
        save=f"seq_batch_{i//batch_size}.pdf",
        show=False
    )
```
